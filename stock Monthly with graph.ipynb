```python
import pandas as pd
import yfinance as yf
import datetime
import matplotlib.pyplot as plt
from datetime import date, timedelta
today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1                      #today's date in format (yyyy - mm - dd)
d2 = date.today() - timedelta(days=360)
d2 = d2.strftime("%Y-%m-%d")       #date of one year back in format (yyyy - mm - dd)
start_date = d2

ydata = yf.download('META', 
                      start=start_date, 
                      end=end_date,interval='1mo') #getting data of specific time period
print(ydata)
```

```python
Ndata = (ydata[['Open','Close']]) #creating new table with required columns only
Ndata['Gain'] = Ndata['Close'] - Ndata['Open'] #creating new column
print(Ndata)
```

```python
plt.plot(Ndata['Close'], label='Price')
plt.ylabel('Price change')
plt.xlabel('Month')
plt.legend()
plt.show()
```

```python
plt.plot(Ndata['Gain'], label='Gain')
plt.ylabel('Price change Monthly')
plt.xlabel('Month')
plt.legend()
plt.show()
```
